```python
#Drop the uncessary columns
#  #   Column                Non-Null Count  Dtype  
# ---  ------                --------------  -----  
#  0   budget                4809 non-null   int64   Drop, not good for the content based
#  1   genres                4809 non-null   object  
#  2   homepage              1713 non-null   object  Drop, not good for the content based
#  3   id                    4809 non-null   int64  
#  4   keywords              4809 non-null   object 
#  5   original_language     4809 non-null   object  Drop, data is baised itself
#  6   original_title        4809 non-null   object  Drop, as the data_values in the native langauges
#  7   overview              4806 non-null   object 
#  8   popularity            4809 non-null   float64 Drop
#  9   production_companies  4809 non-null   object  Drop
#  10  production_countries  4809 non-null   object  Drop
#  11  release_date          4808 non-null   object  Drop
#  12  revenue               4809 non-null   int64   Drop
#  13  runtime               4807 non-null   float64 Drop
#  14  spoken_languages      4809 non-null   object  Drop
#  15  status                4809 non-null   object  Drop
#  16  tagline               3965 non-null   object  Drop, overexagratted
#  17  title                 4809 non-null   object  Drop
#  18  vote_average          4809 non-null   float64 Drop   
#  19  vote_count            4809 non-null   int64   Drop
#  20  movie_id              4809 non-null   int64   Drop
#  21  cast                  4809 non-null   object 
#  22  crew                  4809 non-null   object 


movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
movies['genres']=movies['genres'].apply(lambda obj: list(i['name'] for i in ast.literal_eval(obj)))
movies['keywords']=movies['keywords'].apply(lambda obj: list(i['name'] for i in ast.literal_eval(obj)))
movies['cast']=movies['cast'].apply(lambda obj: list(i['name'] for i in ast.literal_eval(obj)[:3]))
movies['crew']=movies['crew'].apply(lambda obj: list(i['name']  for i in ast.literal_eval(obj) if i['job']=='Director'))
movies['overview'] = movies['overview'].apply(lambda obj:obj.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
print(new_df['tags'][0])
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda obj: ps.stem(obj))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
np.set_printoptions(threshold=1000)
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity  = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
